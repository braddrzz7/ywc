```python
family['race time']=[20.69,21.68,19.84,26.7]
```

```python
race=family.sort_values('race time')
pievalues=race['dumplings'].tolist()
plotlabels=race.index.values.tolist()
barvalues=race['race time'].tolist()
```

```python
# fig1, ax1 = plt.subplots()
figure, (ax1, ax2) = plt.subplots(1, 2, sharey=False, figsize=(14,7))
ax1.bar(np.arange(len(barvalues)),barvalues,color=['aqua','indigo','chocolate','fuchsia'])
ax1.set_ylabel("Time (seconds)");
ax1.set_title("Time to run 3 laps around the house");
ax1.set_xticks(np.arange(len(barvalues)))
ax1.set_xticklabels(plotlabels)
ax2.pie(pievalues, labels=plotlabels,colors=['darkorchid','salmon','lavender','springgreen'])
ax2.set_title('Maximum Number of Dumplings Eaten')
figure.tight_layout(pad=10.0)
display(family); print('')
plt.show()
```
