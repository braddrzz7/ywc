# Data Tables Intro

In this lesson we will learn about data <b>Tables</b>
<ul>
    <li> Data tables are made of <b>rows</b> and <b>columns</b></li>
    <li> We can <b>sort</b> and <b>filter</b> the data to make it easier to read</li>
    <li> We can use <b>pie charts</b> and <b>bar charts</b> to visualize the data</li>
</ul>

```python
import numpy as np
import pandas as pd
```

### Planets

```python
# https://gist.github.com/fusion0202/23624d74061c4236587447f4a97761af
planets = pd.read_csv('https://gist.githubusercontent.com/fusion0202/23624d74061c4236587447f4a97761af/raw/75c80179590249c7bd3c8f966c5d2b2d221b6c64/planets.csv')
planets = planets.rename(columns=lambda x: x.strip());
```

This is called a <b>table</b>.

```python
planets
```

It is made up of <b>columns</b> and <b>rows</b>.<br>
Each row tells us about a different planet.<br>
Each column holds a different piece of information about the planets

Often times we want to rearrange the rows in the table to make it easier to learn things from the data.<br>
We call this <b>sorting</b> the data.<br>
Let's sort this planetary data by number of satellites (moons) to easily see which planet has the most moons

```python
planets.sort_values('Satellites',ascending=False)
```

Now it is very easy to see that Saturn has the most moons, with 64.

Sometimes if the table has too much information it can be harder to read.<br>
Let's reduce the number of columns by removing the values we don't care about.<br>
We can also change the order of the columns.

```python
planets2 = planets.filter(items=['Name','Satellites','Diameter','Mass'])
planets2
```

That is much easer to read.

Suppose we also don't care about all of the planets.<br>
Suppose we only care about planets with at least 5 moons.<br>
We can also remove rows that we don't care about.<br>
We call this <b>filtering</b> the data.

```python
planets3 = planets2.query('Satellites>=5')
planets3
```

### Pokemon

The planetary table is pretty easy to read because it's so small.<br>
It only has 9 rows (one for each planet) and 11 columns.<br>
Let's look at a much larger data table.

```python
# https://www.kaggle.com/mariotormo/complete-pokemon-dataset-updated-090420/version/4?select=pokedex_%28Update_05.20%29.csv
pk=pd.read_csv('pokedex_(Update_05.20).csv')
pk
```

This table has over 1,000 rows and over 50 columns -- too many to see all at once.<br>
Now you can see why it is useful to filter (remove some rows and columns) to make things easier to read.

First let's remove a bunch of columns we don't care about (like the german names of the pokemon)

```python
pk_trimmed = pk.filter(items=['pokedex_number','name','generation','type_1','type_2','height_m','weight_kg','speed','hp','attack','defense','status'])
pk_trimmed
```

Great.  Next let's filter the data to view only the legendary pokemon

```python
pk_trimmed['status'].value_counts()
```

```python
pk_trimmed.loc[pk_trimmed['status'] == 'Sub Legendary']
```

```python
kalos=pk_trimmed.loc[pk_trimmed['generation'] == 6]
kalosLeg=kalos.loc[kalos['status'].isin(['Legendary','Sub Legendary','Mythical'])]
kalosLeg
```

```python
pk_leg = pk_trimmed.loc[pk_trimmed['status'] == 'Legendary']
pk_leg
```

```python
len(pk_leg)
```

Now instead of 1,000 rows we only have 39 rows.<br>
But it's still a lot of rows and I can't see them all at once.

Suppose now we want to find the legendary with the most HP.<br>
Let's sort our filtered table by HP

```python
pk_leg.sort_values('speed',ascending=True)
```

Now it is easy to see that there are two legendaries with 150 HP.

Let's filter the data again!<br>
This time we will remove all pokemon except those from Unova.<br>
Then we will sort by Attack.

```python
pk_leg.loc[pk_leg['generation']==5].sort_values('attack',ascending=False)
```

```python
pk['type_1'].value_counts().plot.pie(figsize=(8, 8))
pk['type_1'].value_counts()
```

```python
pk['generation'].value_counts().plot.pie(figsize=(5, 5))
pk['generation'].value_counts()
```

```python
pk['generation'].value_counts().plot.bar(figsize=(5, 5))
```

```python
pk['name'].apply(lambda s: s[0]).value_counts().plot.bar(figsize=(10, 6))
pk['name'].apply(lambda s: s[0]).value_counts()
```
