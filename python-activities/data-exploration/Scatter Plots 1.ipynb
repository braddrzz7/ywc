#### Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib
matplotlib.rcParams.update({'font.size': 16})
import matplotlib.colors as mcolors

colors = list(mcolors.TABLEAU_COLORS.keys())

def heightAgePlot(x,y,labels=None,s=100,title=None):
    fig,ax = plt.subplots(figsize=(7,4.5))
    if str(y[0]).replace('.','').isnumeric():
        ax.scatter(x,y,s=s)
    else:
        for i in range(len(y)):
            ax.scatter(x[i],y[i],s=s,c=colors[i], label=(labels[i] if labels is not None else None))
    plt.xlabel('Age (years)')
    plt.ylabel('Height (feet)')
    plt.title(title)
    if labels is not None:
        plt.legend(loc=(1.1,0.4))
    plt.grid(True)
```

## <span style='background:yellow'> !!! <font color='red'>Scatter</font> <font color='purple'>Plots</font> !!!  </span>

```python
heightAgePlot([[11],[12],[13],[14],[15]],[[5.5],[4],[6],[5.1],[5.3]],
              labels=['Alice','Bob','Claire','Dan','Ellen'],title='Heights and Ages of my Friends')
```

<b>What does this plot show?</b>Heights and ages for five people.<br>
<b>How old is Dan?</b>14 years old<br>
<b>How tall is Claire?</b>6 feet<br>

### Sometimes the plots show an obvious relationship between the two quantities being plotted

```python
heightAgePlot(np.linspace(9,12,15), np.linspace(4,5,15),title='My Height')
```

#### What does this plot show, and what does it tell us about the relationship between height and age?

One person's height as they get older.  As they get older, they also get taller.  The points make a straight line.  

```python
heightAgePlot(np.linspace(70,90,15), np.linspace(6,5,15),title="Great Grandpa's Height")
```

#### What does this plot show, and what does it tell us?

It's plotting great grandpa's height as he gets older.  As he gets older, his height goes down in a straight line.

```python
ages = np.linspace(9,18,20)
heights = 4 + 2*np.tanh(.3*(ages-9))
heightAgePlot(ages,heights,title='My Height')
```

#### What does this plot show, and what does it tell us?

This plot shows my height as I get older.  It goes up to 18 years, instead of 12 like the earlier plot.  When you get past 12 years you don't grow as much.  Eventually you stop growing at 6 feet.  These points are not straight.  They're curved like a hill.

```python
agesBob = np.linspace(9,19,20)
heightsBob = 4.5 + np.tanh(.3*(ages-9))
heightAgePlot([ages,agesBob],[heights,heightsBob],labels=['Alice','Bob'],s=50)
```

#### What does this plot show, and what does it tell us?

This plot shows how Alice's and Bob's heights change as they get older.  At first Bob is taller than Alice.  But then around age 10 and a half Alice becomes taller and stays taller.  Alice stop growing at 6 feet, and Bob stops growing at 5 and a half feet.

### Sometimes the data clusters in different places on the plot, which can tell us things

```python
steveFamily = np.random.multivariate_normal([25,6.0], [[2,0],[0,.025]], 30)
tinaFamily = np.random.multivariate_normal([35,5.3],[[2,0],[0,.025]], 30)
heightAgePlot([steveFamily[:,0],tinaFamily[:,0]],[steveFamily[:,1],tinaFamily[:,1]],
              labels=["Steve's Family","Tina's Family"],title='Height of Family Members')
```

<b>What does this plot show?  What it tell us about the two families?</b>

It shows height and age of Steve's family and Tina's family.  Each dot represents a different person.  Steve's family is taller than Tina's family.  But Tina's family is older than Steve's family.

### If we split each direction into "big" and "small", we can think of 4 extreme regions in the plot

```python
ursulaFamily = np.random.multivariate_normal([35,6.0], [[2,0],[0,.025]], 30)
valFamily = np.random.multivariate_normal([25,5.3],[[2,0],[0,.025]], 30)
heightAgePlot([steveFamily[:,0],tinaFamily[:,0],ursulaFamily[:,0],valFamily[:,0]],
              [steveFamily[:,1],tinaFamily[:,1],ursulaFamily[:,1],valFamily[:,1]],s=30,
              labels=["Steve's Family","Tina's Family","Ursula's Family","Val's Family"],title='Height of Family Members')
plt.plot([20,39],[5.65,5.65],'k',linewidth=5);
plt.plot([30,30],[4.7,6.4],'k',linewidth=5);
```

<b>What it tell us about the families?</b><br>
Steve's family is tall and young.<br>
Val's family is short and young.<br>
Ursula's family is tall and old.<br>
Tina's family is short and old.
