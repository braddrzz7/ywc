#### Setup

```python
import pandas as pd
import random
import math
df = pd.read_csv('https://corgis-edu.github.io/corgis/datasets/csv/county_demographics/county_demographics.csv',encoding= 'unicode_escape')
```

```python
def barChartAges(state,sortby=None):
    dfstate=df.loc[df['State']==state].sort_values('Age.Percent Under 18 Years')
    step = math.ceil(len(dfstate)/18)
    dfstate = dfstate.iloc[::step, :]
    dfstate.loc[:,'County'] = dfstate.loc[:,'County'].str.replace(' County','')
    dfstate = dfstate.rename(columns={'Age.Percent 65 and Older':'65 and Older',
                           'Age.Percent Under 18 Years':'Under 18'})    
    dfstate['18 to 64'] = 100 - dfstate['65 and Older'] - dfstate['Under 18']
    dfstate = dfstate.sample(frac=1);
    ax = dfstate.plot.bar(x='County',y=['Under 18','18 to 64','65 and Older'],
                 figsize=(12,6),stacked=True,ylabel='Percent',title=state+' Population Age Distribution by County');
    ax.legend(bbox_to_anchor=(1, 1))
    ax.minorticks_on()
    ax.set_axisbelow(True)
    ax.set_ylim((0,100))
    ax.set_xticks((),minor=True)
    ax.tick_params(axis='x', rotation=45);
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='gray',axis='y')
    ax.grid(which='major',axis='y')
```

## Washington

```python
barChartAges('WA')
```
