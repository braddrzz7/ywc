```python
import pandas as pd
```

```python
people = ['dad','mom','charles','alice']
color = ['yellow','red','purple','pink']
age = [38,37,6,3]
```

```python
pd.DataFrame(zip(people,color))
```

```python
df=pd.DataFrame(zip(people,age))
df=df.rename(columns={0:'Name',1:'Age'})
df=df.set_index('Name')
display(df)
df.plot.pie(y='Age',figsize=(6,6));
```

```python
def piechart(labels,values):
    df = pd.DataFrame(zip(labels,values)).set_index(0)
    df.plot.pie(y=1,figsize=(6,6))
def barchart(labels,values):
    df = pd.DataFrame(zip(labels,values)).set_index(0)
    df.plot.bar(y=1,figsize=(6,6))
```

```python
piechart(people,age)
```

```python
names = ['a','b','c','d']
numbers = [1,2,3,4]
barchart(names,numbers)
```

```python
names = ['dad','mom','charles','alice']
ages = [38,37,6,3]
colors = ['yellow','orange','fuchsia','pink']
foods = ['salad','pizza','rice','meatballs']
family=pd.DataFrame(zip(names,ages,colors,foods))
family = family.rename(columns={0:'Name',1:'Age',2:'Color',3:'Food'})
family
```
