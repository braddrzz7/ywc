```python
import matplotlib.pyplot as plt
import numpy as np


dataObject = {'Nurse':[10,90],'Mathematician':[74,26],'Vet':[37,63],
                  'Athlete':[66,34],'Librarian':[16,84],'Photographer':[52,48],
                  'Barber':[88,12],
                  };
allVals = list(dataObject.values());
transposedValues = np.transpose(np.array(allVals));
plt.figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')
barWidth = 0.5;
xpos = np.arange(len(allVals));
labels = ['Men','Women']
for i in range(len(transposedValues)):
    bottoms = np.sum(transposedValues[0:i],axis=0)
    plt.bar(xpos, transposedValues[i], barWidth, align='center', 
        alpha=0.5, label=labels[i],
        bottom = bottoms )
    for j in range(len(xpos)):
        plt.text(xpos[j],bottoms[j]+transposedValues[i][j]/2,str(transposedValues[i][j])+'%',
        ha="center", va="center", color="black",fontsize=12)
plt.xticks(xpos, list(dataObject.keys()),rotation='vertical')
plt.ylabel('Percent')
plt.title('Job Popularity by Sex')
plt.legend(loc='upper center', bbox_to_anchor=(0.9, 1.2),shadow=True, ncol=1)
plt.margins(y=0)
plt.show()
```
